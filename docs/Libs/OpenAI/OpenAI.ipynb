# OpenAI

```python
from openai import OpenAI

API_KEY = ''
BASE_URL = ''

client = OpenAI(api_key=API_KEY, base_url=BASE_URL)

with client.chat.completions.create(
    model=''
    message=[
        {'role': 'system', 'content': ''},
        {'role': 'user', 'content': 'Hello, how are you?'}
    ]
) as response:
    print(response.choices[0].message.content)
```
