# 爬取图片

```python
import requests

# Define the URL of the image to download
img_url = "https://hellorfimg.zcool.cn/provider_image/large/2238428406.jpg"

# Set custom headers to mimic a user agent
headers = {
    "User-Agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Mobile Safari/537.36",
}

try:
    # Send an HTTP GET request to the URL with custom headers
    response = requests.get(img_url, headers=headers)
    response.raise_for_status()  # Check for any request errors
except requests.exceptions.RequestException as e:
    print(f"Request failed: {e}")
else:
    # If the request is successful, save the image data to a file
    with open("img/result.jpg", "wb") as file:
        file.write(response.content)
    print("Image saved successfully")
```
