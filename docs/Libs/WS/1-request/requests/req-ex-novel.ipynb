# 爬取小说

爬取小说《碧甃沉》的第一章。

- `url`: https://www.jjwxc.net/onebook.php?novelid=31816

```python
import requests
from bs4 import BeautifulSoup
from lxml import etree

def make_request():
    # Define the base URL
    base_url = "https://www.jjwxc.net/onebook.php"

    # Define the query parameters as a dictionary
    params = {
        "novelid": "31816",
        "chapterid": "2"
    }

    try:
        # Send a GET request with query parameters
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # Check for request errors

        # Detect and set the encoding based on the content
        response.encoding = response.apparent_encoding  # GB18030
        
        return response
    
    except requests.exceptions.RequestException as e:
        print(f"Request failed: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")
        
def parse_response_by_lxml(response):
    # Parse the HTML content using lxml
    html_tree = etree.HTML(response.text)
    
    # Extract chapter name and content
    chapter_name = html_tree.xpath("//*[@id='oneboolt']/div[2]/div/div[2]/h2/text()")
    content = html_tree.xpath("//*[@id='oneboolt']/div[2]/div/text()")
    
    # Create a dictionary with chapter name as the key and content as the value
    res = {chapter_name[0]: content}
    return res

def parse_response_by_bs(response):
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the element containing the novel content
    novel_content = soup.find('div', class_='novelbody')

    if novel_content:
        # Check the encoding of the web content
        encoding = response.encoding if response.encoding else 'utf-8'

        # Decode and print the content to the console
        print(novel_content.text)
        return novel_content.text
    else:
        print("No novel content found")
    
        
# Make a request to the web page
response = make_request()

# Parse the response using lxml and print the result
res = parse_response_by_lxml(response)
print(res)
```
