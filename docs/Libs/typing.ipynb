# Typing

## 1. Why Do We Need `typing`?

Python is a **dynamically typed** language:

````python
x = 10
x = "hello" # no error
````

This flexibility is powerful, but it can cause problems:

* Harder to understand othersâ€™ code
* Bugs appear only at runtime
* IDEs cannot give accurate auto-completion
* Large projects become difficult to maintain

**Type hints** (introduced in Python 3.5 via PEP 484) help solve these problems.

ðŸ‘‰ Important:

* Type hints are **not enforced at runtime**
* They are used by:

  * IDEs (PyCharm, VS Code)
  * Type checkers (mypy, pyright)
  * Readers of the code

---

## 2. Your First Type Hint

### Without typing

````python
def add(a, b):
    return a + b
````

### With type hints

````python
def add(a: int, b: int) -> int:
    return a + b
````

Explanation:

* `a: int` â†’ parameter type
* `-> int` â†’ return type

Python will still run even if you pass wrong types:

```python
add("a", "b")   # runtime OK, but logically wrong
```

A **type checker** would report an error.

---

## 3. Built-in Types You Can Use Directly

You do **not** need `typing` for simple types:

```python
x: int = 10
y: float = 3.14
name: str = "Alice"
flag: bool = True
```

---

## 4. Why the `typing` Module Exists

Some types cannot be expressed clearly using built-in syntax alone:

* List of integers
* Dictionary with fixed key/value types
* Optional values
* Union types
* Callable functions
* Custom structured data

This is where `typing` comes in.

```python
from typing import List, Dict, Tuple
```

---

## 5. List, Tuple, and Dict

### List

````python
from typing import List

numbers: List[int] = [1, 2, 3]
names: List[str] = ["Alice", "Bob"]
````

âŒ Bad:

````python
numbers: List[int] = [1, "two", 3]
````

### Tuple

Tuples can be **fixed-length**:

````python
from typing import Tuple

point: Tuple[int, int] = (3, 5)
person: Tuple[str, int] = ("Alice", 20)
````

### Dict

````python
from typing import Dict

scores: Dict[str, int] = {
    "Alice": 90,
    "Bob": 85,
}
````

Meaning:

* keys are `str`
* values are `int`

---

## 6. Optional Values (`Optional`)

Sometimes a variable **may be `None`**.

### Without `Optional`

```python
def get_user_name(user_id: int) -> str:
    ...
```

What if no user exists?

### With `Optional`

````python
from typing import Optional

def get_user_name(user_id: int) -> Optional[str]:
    return None
````

Equivalent to:

```python
str | None   # Python 3.10+
```

---

## 7. Union Types (`Union`)

When a value can have **multiple possible types**:

```python
from typing import Union

def parse_id(value: Union[int, str]) -> int:
    return int(value)
```

Python 3.10+ syntax:

```python
def parse_id(value: int | str) -> int:
    return int(value)
```

## 8. Any â€“ â€œDisable Type Checkingâ€

```python

def print_value(value: Any) -> None:
    print(value)
```

`Any` means:

* Accept any type
* Do not perform type checking

âš ï¸ Use sparingly. Overusing `Any` defeats the purpose of typing.

## 9. Functions as Parameters (`Callable`)

```python
from typing import Callable

def apply_func(func: Callable[[int, int], int], a: int, b: int) -> int:
    return func(a, b)
```

Meaning:

* `func` is a function
* Takes two `int` arguments
* Returns an `int`

---

## 10. Custom Data Structures with `TypedDict`

Useful when working with **JSON-like data**.

```python
from typing import TypedDict

class User(TypedDict):
    id: int
    name: str
    active: bool

user: User = {
    "id": 1,
    "name": "Alice",
    "active": True,
}
```

Benefits:

* Clear structure
* IDE auto-completion
* Safer dictionary access

---

## 11. Type Aliases

Improve readability:

```python
from typing import List

UserId = int
Scores = List[int]

def get_scores(user_id: UserId) -> Scores:
    return [90, 85, 88]
```

---

## 12. Generics (`TypeVar`)

When writing **reusable functions**:

```python
from typing import TypeVar, List

T = TypeVar("T")

def first_item(items: List[T]) -> T:
    return items[0]
```

This works for:

* `List[int]`
* `List[str]`
* `List[User]`

## 13. Common Beginner Mistakes

### âŒ Forgetting return type

```python

    return a * 2
```

âœ… Better:

```python

    return a * 2
```

---

### âŒ Using typing everywhere blindly

Typing should improve **clarity**, not add noise.

Bad:

```python
x: int = 1
```

Acceptable:

```python
def process(age: int) -> bool:
    ...
```

---

## 14. Do Type Hints Affect Performance?

No.

* Python ignores type hints at runtime
* Zero performance cost
* Purely for tooling and readability

---

## 15. When Should Beginners Use `typing`?

Recommended usage:

* Function parameters and return values
* Public APIs
* Data structures (list/dict)
* Team projects
* Medium to large codebases

Optional usage:

* Small scripts
* One-off experiments
